Data from the World Bank Indicators portal, which is an incredibly rich resource. Contains the following fields:
    
- iso2c iso3c Two- and Three-letter codes for each country, assigned by the International Organization for Standardization.
- country Country name.
- year population Estimated total population at mid-year, including all residents apart from refugees.
- gdp_percap Gross Domestic Product per capita in current international dollars, corrected for purchasing power in different territories.
- life_expect Life expectancy at birth, in years.
- population Estimated total population at mid-year, including all residents apart from refugees.
- birth_rate Live births during the year per 1,000 people, based on mid-year population estimate.
- neonat_mortal_rate Neonatal mortality rate: babies dying before reaching 28 days of age, per 1,000 live births in a given year.
- region income World Bank regions and income groups,

__Replace values with the median value of column:__

```python
data['gdp_percap'].fillna(data['gdp_percap'].median(),inplace=True)
```

```python
data.head()
```

__ find the mean gdp_percap for each country. __

```python
data.groupby(['country']).mean()
```

```python
avg_gdp_percap = data.groupby(['country']).mean()['gdp_percap']
avg_gdp_percap
```

```python
top_five_countries = avg_gdp_percap.sort_values(ascending=False).head()
top_five_countries
```

```python
united = data[data['country'] == 'United Arab Emirates']
united
```

```python
plt.plot(united['year'], united['gdp_percap'])
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('GDP Per capita of Macao SAR, China')
' the line plot doesnâ€™t give us a good representation that we can make meaning from, so letâ€™s try and visualize it in a barplot to get a better understanding of the data:'
```

```python
united.plot.bar(x = 'year', y = 'gdp_percap')
```

```python
plt.subplot(311)
plt.title('GDP Per Capita')
plt.plot(united['year'], united['gdp_percap'])
plt.subplot(312)
plt.title('GDP in Billions')
plt.plot(united['year'], (united['population']*united['gdp_percap']/10**9))
plt.subplot(313)
plt.title('Population in Millions')
plt.plot(united['year'], united['population']/10**6)
plt.tight_layout()
```
